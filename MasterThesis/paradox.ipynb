```python
from qiskit import QuantumCircuit
from qiskit.primitives import Sampler, Estimator
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.draw()

qc.measure_all()

result = Sampler().run(qc).result()
plot_histogram(result.quasi_dists)
```

```python
qc =QuantumCircuit(2,1)
qc.h([0,1])
qc.cz(0,1)
qc.h(0)

qc.measure(0,0)
display(qc.draw())
result = Sampler().run(qc).result().quasi_dists
resultaer = AerSimulator().run(qc).result().get_counts()
display(plot_histogram(result))
plot_histogram(resultaer)
```
